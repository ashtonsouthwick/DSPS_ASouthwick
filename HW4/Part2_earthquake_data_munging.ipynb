````python
# your code here
ss = []

# loop through all year pairs as year and fraction of year
for t in [(1984, 1986.5),
          (1990.3, 1992.1),
          (1994.6, 1995.6),
          (1996.1,1996.5),
          (1997,1997.6),
          (1997.75,1998.15),
          (1998.25,1999.35),
          (2000.55,2000.8),
          (2000.9 ,2001.25),
          (2001.6 ,2002),
          (2002.5 ,2003)]:
    # for each pair I convert the year to datetime (easy)
    # then I convert the fraction of year and convert it to days and convert that to a delta-time
    # finally I add that deltatime to the year datetime.
    ss.append((pd.to_datetime(int(t[0]), format="%Y")
               + pd.Timedelta(days=(t[0] - int(t[0])) * 365.25),
               pd.to_datetime(int(t[1]), format="%Y")
               + pd.Timedelta(days=(t[1] - int(t[1])) * 365.25)))

print ("the {} timestamp pairs that are the boundaries of good data periods are:".format(len(ss)))
# data between start and end of first time period
gooddates = pd.Series(
    np.any(
        [(earthquakes.newdatetime >= start) & (earthquakes.newdatetime <= end) for start, end in ss],
        axis=0
    ),
    index=earthquakes.index,
    name="newdatetime"
)
gooddates
````

For each of the 11 valid periods I create a condition datetime >= period_start AND datetime <= period_end where AND is a product of booleans: True * False = False.

I want to take all chinks of data that satisfy this condition for any of the 11 chunks of time so that is a logical OR operator, or a sum of Booleans: True + False = True

This is because True really corresponds to "not 0" and False to "0". So "not 0" * 0 = 0 and "not 0" + 0 = not 0

````python
# your code here
# start with all False
gooddates = np.zeros(len(earthquakes), dtype=bool)

# loop through each stationary interval and OR them together
for start, end in ss:
    gooddates = gooddates + ((earthquakes.newdatetime >= start) * (earthquakes.newdatetime <= end))

print ("there are {} earthquakes falling in the selected stationary periods".
       format(gooddates.sum()))
````

### important: the paper does not do it but I want you to formally choose a threshold before you do this test. Choose a p-value (=$\alpha$ thresold: call your variable ```alpha```) of 3-$\sigma$.

Follow the instructions (algorithm) in **Section 3 paragraph 3** and compare your results with the results in table 1 with a threshold  of = 0.01 and 0.001

Do it for all 5 magnitude thresholds as indicated in the paper (and in Table 1).

Note that the x axis in plot Fig 1 is in _log space_. Use ```np.log10()``` to take the logarithm of the time gaps.

The pseudocode for the algorithm is [here](https://github.com/fedhere/DSPS_FBianco/blob/master/HW3/Corral2018pseudocodeKStest.md).


Reproduce the paper Fig 1 and Table 1. In the Table report the size of each dataset after cleaning the value of the statistic and the p-value, as done in Corral 2018. Use the scipy function for the 2 sample KS test. (resources [here](https://colab.research.google.com/notebooks/markdown_guide.ipynb#scrollTo=70pYkR9LiOV0) to learn about the table syntax in markdown)

````python
#Used chatgpt here... sorry, dont remember prompt, it was a lot of back and forth

mag_thresholds = [2.0, 2.5, 3.0, 3.5, 4.0]
cutoffs = [0.01, 0.001] # m vals
alpha_3sigma = 0.0027 # ~3-sigma two-sided approx
alphas_to_check = [0.01, 0.001, alpha_3sigma]

def iterative_rescale_and_trim(taus_seconds, m_cut, max_iter=10):
    taus = np.array(taus_seconds, copy=True)
    if len(taus) == 0:
        return np.array([]), np.nan, np.array([])
    for it in range(max_iter):
        R = 1.0 / np.mean(taus)
        x = R * taus
        keep_mask = x >= m_cut
        if keep_mask.all():
            break
        taus = taus[keep_mask]
        if len(taus) == 0:
            break
    if len(taus) == 0:
        return np.array([]), R, taus
    R = 1.0 / np.mean(taus)
    x = R * taus
    return x, R, taus

all_results = {}

for m_cut in cutoffs:
    rescaled = {} # rescaled distributions per magnitude threshold
    sizes = {} # number of events after trimming
    Rs = {}
    for Mk in mag_thresholds:
        taus = eqClean.loc[eqClean['mag'] >= Mk, 'dt_seconds'].dropna().values
        # run iterative trimming/rescaling (filtering on x)
        x, Rk, final_taus = iterative_rescale_and_trim(taus, m_cut, max_iter=10)
        rescaled[Mk] = x
        sizes[Mk] = len(final_taus)
        Rs[Mk] = Rk

    # compute KS stat & p-value matrices
    n = len(mag_thresholds)
    dmat = np.zeros((n, n))
    pmat = np.ones((n, n))
    for i, Mi in enumerate(mag_thresholds):
        for j, Mj in enumerate(mag_thresholds):
            if i == j:
                dmat[i,j] = 0.0
                pmat[i,j] = 1.0
            else:
                xi = rescaled[Mi]
                xj = rescaled[Mj]
                if len(xi) > 1 and len(xj) > 1:
                    ks = stats.ks_2samp(xi, xj)
                    dmat[i,j] = ks.statistic
                    pmat[i,j] = ks.pvalue
                else:
                    dmat[i,j] = np.nan
                    pmat[i,j] = np.nan

    all_results[m_cut] = {
        'rescaled': rescaled,
        'sizes': sizes,
        'R': Rs,
        'dmat': dmat,
        'pmat': pmat
    }

# producing tabels
for m_cut, out in all_results.items():
    sizes = out['sizes']
    dmat = out['dmat']
    pmat = out['pmat']
    n = len(mag_thresholds)

    mat = np.empty((n, n), dtype=object)
    for i, Mi in enumerate(mag_thresholds):
        for j, Mj in enumerate(mag_thresholds):
            if i == j:
                mat[i,j] = '-'
            elif j > i:
                # upper triangle: p-value in percent with 1 decimal
                if not np.isnan(pmat[i,j]):
                    mat[i,j] = f"{pmat[i,j]*100:.1f}%"
                else:
                    mat[i,j] = 'nan'
            else:
                # lower triangle: KS stat with 3 decimals
                if not np.isnan(dmat[i,j]):
                    mat[i,j] = f"{dmat[i,j]:.3f}"
                else:
                    mat[i,j] = 'nan'

    df = pd.DataFrame(mat, index=[f"M â‰¥ {m}" for m in mag_thresholds],
                          columns=[f"M â‰¥ {m}" for m in mag_thresholds])
    df.insert(1, "N", [sizes[m] for m in mag_thresholds])
    print(f"\nTable for m = {m_cut}")
    print(df.to_markdown())

# Plot the rescaled CDFs like Fig 1 for both cutoffs
for m_cut, out in all_results.items():
    plt.figure(figsize=(8,6))
    for Mk in mag_thresholds:
        x = np.sort(out['rescaled'][Mk])
        if len(x) == 0:
            continue
        y = np.arange(1, len(x)+1) / len(x)
        plt.plot(np.log10(x), y, marker='.', linestyle='none', label=f"M â‰¥ {Mk}")
    plt.xlabel(r"$\log_{10}(x = R_k \tau)$")
    plt.ylabel(r"$P(x)$ (empirical CDF)")
    plt.title(f"Rescaled CDFs (m_cut = {m_cut})")
    plt.legend()
    plt.grid(True, which='both', ls='--', alpha=0.5)
    plt.show()
````
